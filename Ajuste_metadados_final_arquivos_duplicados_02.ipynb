# Ajuste de notícias duplicadas na base da hemeroteca de PEB

## Ver entradas duplicadas

```python
dataset["nome_arquivo_pdf_lower"] = dataset["nome_arquivo_pdf"].str.lower()
dataset["nome_arquivo_pdf_lower"] 
```

```python
duplicadas = dataset[dataset["nome_arquivo_pdf_lower"].duplicated(keep=False)]
duplicadas.shape
```

```python
#Comparação
duplicadas
```

```python
#Seleção das notícias que vamos manter
duplicadas["contar_caixa_alta"] = duplicadas["nome_arquivo_pdf"].str.count(r"[A-Z]")
mais_caixa_alta = duplicadas.groupby("nome_arquivo_pdf_lower")["contar_caixa_alta"].idxmax()
selecionadas = duplicadas.loc[mais_caixa_alta]
selecionadas
```

## Seleção de uma das entradas no json
* Todos com caixa baixa, por exemplo

```python
nao_selecionadas = duplicadas[~duplicadas.index.isin(selecionadas.index)]
nao_selecionadas.shape
```

```python
nao_selecionadas
```

```python
nao_selecionadas["caminho_pdf"] = nao_selecionadas["dir_pdf"] + "/" + nao_selecionadas["nome_arquivo_pdf"]
nao_selecionadas["caminho_pdf"]
```

```python
caminho_pdf_excluir = nao_selecionadas["caminho_pdf"].tolist()
caminho_pdf_excluir
```

```python
#percorre as entradas e dentro delas o diretório e a lista de tifs
nao_selecionadas["caminho_arquivo_tif"] = nao_selecionadas.apply(
    lambda linha: [linha["dir_arquivo"]+tif for tif in linha["nome_arquivo_tif"]], axis=1)

#aplica o caminho em todos os arquivos tif
nao_selecionadas = nao_selecionadas.explode("caminho_arquivo_tif")

#cria um novo dataset
nao_selecionadas = nao_selecionadas[["dir_arquivo", "caminho_arquivo_tif"]]

#lista os caminhos tif
lista_caminhos_tif = nao_selecionadas["caminho_arquivo_tif"].tolist()
lista_caminhos_tif
```

```python
#saltar as não selecionadas para verificar a validade do código
selecionadas = dataset.drop(nao_selecionadas.index)
selecionadas.shape
```
