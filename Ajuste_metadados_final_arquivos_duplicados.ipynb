## Ver entradas duplicadas

```python
nome_arq = dataset["nome_arquivo_pdf"].apply(lambda x: [type(i) for i in x])
print(nome_arq)
```

```python
df = dataset

# definindo a opção de exibição para mostrar todas as colunas sem cortes
pd.set_option('display.max_colwidth', None)

# convertendo todas as strings para letras minúsculas
df['nome_arquivo_pdf'] = df['nome_arquivo_pdf'].str.lower()

# encontrando linhas duplicadas na coluna 'nome_arquivo_pdf'
duplicados = df[df['nome_arquivo_pdf'].duplicated()]["nome_arquivo_pdf"]

# imprimindo os arquivos duplicados
duplicados
```

```python
# definindo a opção de exibição para mostrar todas as colunas sem cortes
pd.set_option('display.max_colwidth', None)

# convertendo todas as strings para letras minúsculas
df['nome_arquivo_pdf'] = df['nome_arquivo_pdf'].str.lower()

# encontrando linhas duplicadas na coluna 'nome_arquivo_pdf'
duplicados = df[df['nome_arquivo_pdf'].duplicated(keep=False)].sort_values('nome_arquivo_pdf')

# criando um novo DataFrame com as duplicatas lado a lado
df_lado_a_lado = pd.DataFrame(index=duplicados['nome_arquivo_pdf'].unique())
for coluna in duplicados.columns:
    df_lado_a_lado[f'{coluna}_duplicado'] = duplicados.groupby('nome_arquivo_pdf')[coluna].apply(lambda x: '|'.join(x.tolist())).values

# imprimindo os arquivos duplicados lado a lado
print(df_lado_a_lado)
```

```python
df = dataset

# definindo a opção de exibição para mostrar todas as colunas sem cortes
pd.set_option('display.max_colwidth', None)

# convertendo todas as strings para letras minúsculas na coluna 'nome_arquivo_pdf'
df['nome_arquivo_pdf'] = df['nome_arquivo_pdf'].str.lower()

# agrupando os registros com o mesmo nome de arquivo
grupos = df.groupby(['nome_arquivo_pdf'])

# concatenando os valores de cada grupo em uma única string
duplicados = grupos.apply(lambda x: ' | '.join(x['nome_arquivo_pdf']))

# imprimindo os arquivos duplicados lado a lado
duplicados.shape
```
